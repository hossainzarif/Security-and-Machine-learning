```
def x_or(a,b):
  x_or_res=[]
 
  for i in range (0,len(a)):
    if (a[i]==b[i]):
      x_or_res.append(0)
    else:
      x_or_res.append(1)

  return x_or_res
```

```
def hex_to_binary(hex_val):
  hex_bin_res=format(int(hex_val, 16), "024b")
  return hex_bin_res 
```

```

def bin_to_hex(a):
  
  return '%06X' % int(a, 2)

```

```
def dec_to_binary(a):
  bin_res="{0:04b}".format(a)  
  return bin_res
```

```
def left_shift(bin,n):
  li= list(map(int, bin))  
  
  for i in range(0,n):
      k=li[0]
      li.pop(0)
      li.append(k)
     
  #left_res=''.join(map(str, li)) 

  
  return li

```

```
def reshuffle_table(pc_1,ac):
  reshuffled=[]
  for i in range (0,len(pc_1)):
      k= ac[(pc_1[i])-1]
      reshuffled.append(k)
    
  return reshuffled
```

```
def expansion(expansion_table,ac_table):
  expand_reshuffled=[]
  for i in range (0,len(expansion_table)):
      k= ac_table[(expansion_table[i])-1]
      expand_reshuffled.append(k)
    
  return expand_reshuffled


```

```
def s_box_analyze(box,row,column):
  x=box[row]
  exact_value=x[column]

  #res = int("".join(map(str, list))) 
  conv_binary= dec_to_binary(exact_value)

  return list(map(int, conv_binary))  
  
```

```
def sbox_checking(s1_box,s2_box,s3_box,text):
  half_s= int(len(text)/3)
  st_half= text[:half_s]
  nd_half=text[half_s:(2*half_s)]
  rd_half=text[(2*half_s):]

  print(st_half)
  print(nd_half)
  print(rd_half)

  s1_row = binary_to_dec(str(st_half[0])+str(st_half[5]))
  s1_column=binary_to_dec(str(st_half[1])+str(st_half[2])+str(st_half[3])+str(st_half[4]))
  s2_row = binary_to_dec(str(nd_half[0])+str(nd_half[5]))
  s2_column=binary_to_dec(str(nd_half[1])+str(nd_half[2])+str(nd_half[3])+str(nd_half[4]))
  s3_row =binary_to_dec(str(rd_half[0])+str(rd_half[5]))
  s3_column=binary_to_dec(str(rd_half[1])+str(rd_half[2])+str(rd_half[3])+str(rd_half[4]))

 

  s1=s_box_analyze(s1_box, s1_row, s1_column)
  s2=s_box_analyze(s2_box,s2_row, s2_column)
  s3=s_box_analyze(s3_box,s3_row, s3_column)
  
  S=s1+s2+s3
  
  return S
```

**Sub keys Generated for Simplified DES.**

```
def Subkey_generator(initial_key,iteration_table,pc1_table,pc2_table):
  sub_keys=[]
  reshuffled_list=reshuffle_table(pc1_table,initial_key)
  print('Permuted with PC-1 Table:== ',reshuffled_list)

  half= len(reshuffled_list)/2

  l_side= reshuffled_list[:mt.floor(half)]
  r_side=reshuffled_list[mt.ceil(half)-1:]

  print('Divided into Left side:== ',l_side)
  print('Divided into Right side:== ',r_side)

  print('\n')
  for i in range (0,len(iteration_table)):
    iterations= iteration_table[i]
    print('ROUND:=========== '+str(i+1))

    print('before left shift left side',l_side)
    print('before left shift right side',r_side)
    l_side_shift=left_shift(l_side,iterations)
    r_side_shift=left_shift(r_side,iterations)

    k_key=l_side_shift+r_side_shift

    print('Left Side left Shifted for round '+str(i+1)+':==',l_side_shift)
    
    print('Ride Side left Shifted for round '+str(i+1)+':==',r_side_shift)

    print('Merged Left and right sides',k_key)

 
    keys_gen= reshuffle_table(pc2_table,k_key)
    print('Subkey Generated for round '+str(i+1)+':==',keys_gen)
    sub_keys.append(keys_gen)
    l_side= l_side_shift
    r_side = r_side_shift
    print('\n')
  #print(sub_keys)
  #print(sub_keys[0])

  return sub_keys
  

  
```

**Encryption Processing**

```
def Encryption(init_plain_text,init_key,iteration_table,ip_table,expansion_table,permutation_table,pc1_table,pc2_table,ip_inverse_table,s1_table,s2_table,s3_table):

  binary_key= hex_to_binary(init_key)
  print('Hexadecimal to binary key:== '+binary_key)

  binary_plain_text=hex_to_binary(init_plain_text)
  print('Hexadecimal to binary plain text:== '+binary_plain_text)
  print('\n\n ')
  print('SUBKEY GENERATION PROCESS ONGOING:=================================================================================================================================')
  keys=Subkey_generator(binary_key,iteration_table,pc1_table,pc2_table)

  print('\n\n ')
  print('ENCRYPTION PROCESS ONGOING:=================================================================================================================================')
  IP_text= reshuffle_table(ip_table,binary_plain_text)
  print('Permuted plain text',IP_text)

  for i in range (0,len(iteration_table)):
    print('\n')
    print('ENCRYPTION ROUND:=========== '+str(i+1)) 
    half= int(len(IP_text)/2)
  
    LH= IP_text[:half]
    RH=IP_text[half:]

    print('Left side:== ',LH)

    print('Right side:== ',RH)

    expanded= expansion(expansion_table,RH)
    print('Expanded Right Side:==',expanded)

    expanded_int=list(map(int, expanded)) 
    X_OR= x_or(expanded_int,keys[i])
    print('X_Or with key '+str(i+1) +':===',X_OR)

    S_box_res=sbox_checking(s1_table,s2_table,s3_table,X_OR)

    print('S-box Output:== ',S_box_res)

    P_box= reshuffle_table(permutation_table,S_box_res)

    print('P-box Output:== ',P_box)
    LH_int=list(map(int, LH)) 
    X_OR_wleftside= x_or(P_box,LH_int)

    print('X-OR with left side Output:==',X_OR_wleftside)

    
    RH_int=list(map(int, RH)) 
    result_after_rounds= RH_int+X_OR_wleftside
    print('X-OR with left side and Right side Merged :==',result_after_rounds)
    IP_text=result_after_rounds

  half_cipher= int(len(result_after_rounds)/2)
  
  LH_cipher= IP_text[:half_cipher]
  RH_cipher=IP_text[half_cipher:]

  swap=RH_cipher+LH_cipher
  

  cipher_text= reshuffle_table(ip_inverse_table,swap)

  print('\n')
  print("Gained Cipher Text in Binary :== ",cipher_text)

  left_res=''.join(map(str, cipher_text))
  print('\n')
  print("Gained Cipher Text in Hexa-decimal:== "+bin_to_hex(left_res))
  

  Decrypted_Text= Decryption(cipher_text,keys,iteration_table,ip_table,expansion_table,permutation_table,pc1_table,pc2_table,ip_inverse_table,s1_table,s2_table,s3_table)




 
```
